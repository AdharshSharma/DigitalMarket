```python
choice = int(input("Select an option (1–5): "))

match choice:
    case 1:
        print("Option 1")
    case 2:
        print("Option 2")
    case 3:
        print("Option 3")
    case 4:
        print("Option 4")
    case 5:
        print("Option 5")
    case _:
        print("Invalid choice! Please choose between 1 and 5.")
```
